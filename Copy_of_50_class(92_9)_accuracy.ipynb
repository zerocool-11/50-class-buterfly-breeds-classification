```
from keras.applications import MobileNet
from keras.layers import Input
import PIL
# VGG16 was designed to work on 224 x 224 pixel input images sizes
img_rows = 220
img_cols = 220 

# Re-loads the VGG16 model without the top or FC layers
basemodel = MobileNet(weights = 'imagenet', 
                 include_top = False,
                 input_tensor = Input(shape=(img_rows, img_cols, 3)))

# Here we freeze the last 4 layers 
# Layers are set to trainable as True by default
print(basemodel.summary())
for layer in basemodel.layers:
    layer.trainable = False
# Let's print our layers 
for (i,layer) in enumerate(basemodel.layers):
    print(str(i) + " "+ layer.__class__.__name__, layer.trainable)
```

```
basemodel.input
```

```
from keras.regularizers import l2
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, AveragePooling2D, MaxPooling2D, ZeroPadding2D
from keras.layers.normalization import BatchNormalization
from keras.models import Model
from keras import regularizers 
model = basemodel.output
#model = AveragePooling2D(padding='same')(model)
model = Flatten(name="flatten")(model)
model = Dense(350, activation="relu")(model)
model = Dropout(0.5)(model)
model = Dense(50,activation="softmax")(model)
```

```
modelnew = Model(inputs=basemodel.input, outputs=model)

print(modelnew.summary())
```

```
from keras.preprocessing.image import ImageDataGenerator
from keras.optimizers import Adam
```

```
batch_size=32
inp_size=(220,220)

train_data_dir='DATA/TRAIN/'
train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest',
        validation_split=0.2)
test_datagen = ImageDataGenerator(rescale=1./255)


train_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=inp_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True,
    subset='training') # set as training data

validation_generator = train_datagen.flow_from_directory(
    train_data_dir, # same directory as training data
    target_size=inp_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True,
    subset='validation')

adam = Adam(lr=0.0001)

modelnew.compile(loss = 'categorical_crossentropy',
              optimizer = adam,
              metrics = ['accuracy'])
```

```
import tensorflow as tf



all_callbacks = [
    tf.keras.callbacks.EarlyStopping(monitor = 'loss',
                          min_delta = 0,
                          patience = 5,
                          verbose = 1,
                          restore_best_weights = True),
    tf.keras.callbacks.ModelCheckpoint(filepath='callbacks/model-mobilenet-1dense-layer.{epoch:02d}-{val_loss:.2f}.h5',     
                                  monitor='val_accuracy',
                                  mode='auto',
                                  save_best_only=True,),
    tf.keras.callbacks.TensorBoard(log_dir='callbacks/logs'),
]
```

```


nb_train_samples = 3601
nb_validation_samples = 878
epochs = 15
batch_size = 32

modelnew.fit(
    train_generator,
    steps_per_epoch = nb_train_samples // batch_size,
    epochs = epochs,
    validation_data = validation_generator,
    validation_steps = nb_validation_samples // batch_size,
    callbacks=all_callbacks)

modelnew.save("before_finetune.h5")
```
